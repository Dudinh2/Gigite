```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
num_row = 50

data = {
    'produto':[f'Produto_{i}'for i in range(1,num_row + 1)],
    'Categoria': np.random.choice(['Televisão','Computador','Celular', 'Tablet', ' Impressora'], size=num_row),
    'Preço': np.random.uniform(100.0, 2000.0, size = num_row).round(2),
    'Quantidade_Estoque': np.random.randint(1,50, size= num_row),
    'Posse': np.random
}
```

```python
df = pd.DataFrame(data)
```

```python
df ['Multiplique'] = df['Preço']*df['Quantidade_Estoque']
```

```python
def Multiplicação (x):
    return x ['Preço']* x['Quantidade_Estoque']
```

```python
df["resultado"]= df.apply(Multiplicação, axis=  1)
```
