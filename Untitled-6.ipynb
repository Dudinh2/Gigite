```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
dias = np.arange(1,31)
vendas = np.random.randint(50,150,size=30)
```

```python
df_vendas= pd.DataFrame({'Dias':dias,'Vendas':vendas})
```

```python
df_vendas
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_vendas['Dias'],df_vendas['Vendas'],marker= 'o',linestyle = '-', color='g')
plt.title('Acompanhamento de vendas Dirias', fontsize=16)
plt.xlabel('Dia dos ms',fontsize=14)
plt.xlabel('Dia dos dirias',fontsize=14)
plt.grid(False)
plt.show()
```

```python
categorias = ['Eletronicos','Roupas','Alimentos','Mveis','Livros']
vendas_totais = [1200,850,950,500,750]
df_vendas_categorias = pd.DataFrame({'Categoria':categorias,'Vendas':vendas_totais})
```

```python
df_vendas_categorias
```

```python
plt.figure(figsize=(10,6))
plt.bar(df_vendas_categorias['Categoria'],df_vendas_categorias['Vendas'],color='pink')
plt.title('Vendas Totais por Categorias de Produto - Ms Atual',fontsize=16)
plt.xlabel('Categoria de Produto',fontsize=14)
plt.xlabel('Vendas totais',fontsize=14)
```

```python
marcas = ['Apple','Sansung','Xiaomi','OnePlus',' Outras']
participacao_mercado = [35,30,15,10,10]
```

```python
participacao_mercado
```

```python
df_participacao= pd.DataFrame({'Marca':marcas,'Participacao':participacao_mercado})
```

```python
df_participacao
```

```python
plt.figure(figsize=(8,8))
plt.pie(df_participacao['Participacao'], labels=df_participacao['Marca'],autopct='%1.1f%%',colors=['gold','skyblue','orange','pink','lightgreen'])
plt.title('Participao no mercado de Smartphones',fontsize=16)
plt.show()
```
